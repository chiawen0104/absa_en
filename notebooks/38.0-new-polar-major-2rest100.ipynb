```python
import pandas as pd
import json
```

```python
with open('../data/yelp/ANS2-correct_polar-major_2rest100.json') as f:
    gold = json.load(f)

gold[:2]
```

### Rest1

```python
gold1 = [i for i in gold if i['business_id'] == '_ab50qdWOk0DdB6XOrBitw']
key1 = []

for text in gold1:
    if len(text['opinions']) > 0:
        for op in text['opinions']:
            if len(op['Target_key']) > 0:
                for key in op['Target_key']:
                    if key not in key1:
                        key1.append(key)

len(key1)
```

```python
TRUE_polar_ans1, FALSE_polar_ans1, accur1 = [], [], []
total_opinion1, Pos1, Neg1, TP1, FP1 = [], [], [], [], []
TN1, FN1 = [], []

for k in key1:
    t, f = 0, 0
    opinion, pos, neg, tp, fp = 0, 0, 0, 0, 0

    for g in gold1:
        if (len(g['opinions']) > 0): # non empty 'opinions'
            for o in g['opinions']: # every opinion in 'opinions'
                if (len(o['Target_key']) > 0): # non empty 'Target_key'
                    for tgk in o['Target_key']: # every word in 'Target_key'

                        if tgk == k: 
                            opinion += 1

                            if o['Polarity'] == 'Positive':
                                 pos += 1
                            else: 
                                neg += 1

                            if o['Polar_ans']: # gold ans: True
                                t += 1
                                if o['Polarity'] == 'Positive': 
                                    tp += 1
                            else: # gold ans: False
                                f += 1
                                if o['Polarity'] == 'Positive': 
                                    fp += 1


    total_opinion1.append(opinion)
    TRUE_polar_ans1.append(t)
    FALSE_polar_ans1.append(f)
    accur1.append("%.2f%%" % (100 * t/opinion))
    Pos1.append(pos)
    Neg1.append(neg)
    TP1.append(tp)
    FP1.append(fp)
    TN1.append(t-tp)
    FN1.append(f-fp)
```

```python
# rest1's stat

df_stat1 = pd.DataFrame({'key': key1, 'total_opinion': total_opinion1, 'TRUE_polar_ans': TRUE_polar_ans1, 'FALSE_polar_ans': FALSE_polar_ans1,
                        'accuracy': accur1, 'TP': TP1, 'TN':  TN1, 'FP': FP1, 'FN': FN1,
                        'Positive': Pos1, 'Negative': Neg1})

df_stat1 = df_stat1.sort_values(['total_opinion'], ascending=False)
df_stat1[df_stat1.total_opinion >= 5]
# df_stat1
```

```python
# df to csv
df_stat1.to_csv('../data/yelp/STAT2_polar-major_rest1.csv', index=False)
```
